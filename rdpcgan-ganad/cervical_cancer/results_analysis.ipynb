```python
import os
import pandas as pd

def process_individual_folder(folder_path):
    auc_values = []
    apr_values = []

    # Iterate over each file in the folder
    for i in range(10):
        file_name = f"results-pategan-{i}.csv"
        file_path = os.path.join(folder_path, file_name)
        if os.path.isfile(file_path):
            with open(file_path, 'r') as file:
                lines = file.readlines()
                if len(lines) >= 3:
                    auc_values.append(float(lines[1].strip()))
                    apr_values.append(float(lines[2].strip()))

    # Extract the first word of the folder name
    folder_name = os.path.basename(folder_path)
    # print(folder_name.split("_"))
    first_word = folder_name.split("eps=")[0] + folder_name.split("eps=")[1].split("_")[0]

    # Convert to pandas DataFrame for easy calculations
    df = pd.DataFrame({
        'AUC': auc_values,
        'APR': apr_values
    })

    # Print out the values
    print(f"Folder {first_word})")
    print("AUC Values:")
    print(df['AUC'].to_string(index=False))
    print("\nAPR Values:")
    print(df['APR'].to_string(index=False))

    # Calculate and print statistics
    print("\nStatistics for AUC:")
    print(f"Mean: {df['AUC'].mean()}")
    print(f"Median: {df['AUC'].median()}")
    print(f"Variance: {df['AUC'].var()}")

    print("\nStatistics for APR:")
    print(f"Mean: {df['APR'].mean()}")
    print(f"Median: {df['APR'].median()}")
    print(f"Variance: {df['APR'].var()}")
    print("\n" + "-"*40 + "\n")

def process_results_folder(results_folder):
    # Iterate over each subfolder in the results folder
    for subfolder in os.listdir(results_folder):
        subfolder_path = os.path.join(results_folder, subfolder)
        if os.path.isdir(subfolder_path):
            process_individual_folder(subfolder_path)

# Example usage
results_folder = "results"  # Replace with your actual results folder path
process_results_folder(results_folder)
```
